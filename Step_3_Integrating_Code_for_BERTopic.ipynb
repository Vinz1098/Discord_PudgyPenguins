```python
import pandas as pd

# Read the first sheet
sheet1 = pd.read_excel('Phase 1 (announcement).xlsx')

# Read the second sheet
sheet2 = pd.read_excel('cleaned_discord_huddlemessages_P1.xlsx')

# Convert 'Date' column in sheet2 to datetime64[ns] format
sheet2['Date'] = pd.to_datetime(sheet2['Date'])

# Merge the two sheets based on the 'Date' column
result_df = pd.merge(sheet1, sheet2, how='left', on='Date', suffixes=('_sheet1', '_sheet2'))

# Drop rows with missing values
result_df = result_df.dropna()

# Display the result DataFrame
display(result_df)
```

```python
# Save the cleaned DataFrame to a CSV file
result_file_path = 'merged_and_cleaned Phase1forBertopic_data.csv'
result_df.to_csv(result_file_path, index=False)
```
