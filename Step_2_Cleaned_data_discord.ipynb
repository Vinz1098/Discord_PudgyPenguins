```python
import pandas as pd
from google.colab import files
import re
from nltk.corpus import stopwords
import nltk

# Download NLTK stopwords data
nltk.download('stopwords')

# Load the Excel file into a DataFrame
file_path = 'discord_huddlemessages_jun2021_to_aug2021_with_usersx.xlsx'
df = pd.read_excel(file_path)

# Rename 'Timestamps' column to 'Date'
df.rename(columns={'Timestamps': 'Date'}, inplace=True)

# Convert the 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%dT%H:%M:%S.%f+00:00')

# Extract date in the format dd-mm-year
df['Date'] = df['Date'].dt.strftime('%d-%m-%Y')

# Specify the column you want to clean
column_to_clean = 'Content'  # Replace with the actual column name

# Clean the specified column by removing phrases less than three words
df[column_to_clean] = df[column_to_clean].apply(lambda x: ' '.join([word for word in str(x).split() if len(word) >= 4]))

# Remove duplicates in the specified column
df = df.drop_duplicates(subset=[column_to_clean])

# Remove rows with missing values in the specified column
df = df.dropna(subset=[column_to_clean])

# Remove special characters and keep only alphanumeric characters in the specified column
df[column_to_clean] = df[column_to_clean].apply(lambda x: re.sub(r'[^a-zA-Z0-9\s]', '', str(x)))

# Remove stop words in the specified column
stop_words = set(stopwords.words('english'))
df[column_to_clean] = df[column_to_clean].apply(lambda x: ' '.join([word for word in str(x).split() if word.lower() not in stop_words]))

# Remove rows where the specified column has only numbers
df = df.loc[~df[column_to_clean].str.isnumeric()]

# Save the modified DataFrame to a new Excel file
cleaned_file_path = 'cleaned_discord_huddlemessages_P1.xlsx'
df.to_excel(cleaned_file_path, index=False)

# Display the first few rows of the cleaned DataFrame
print(df.head())

# Download the cleaned file
files.download(cleaned_file_path)
```
