```python
from google.colab import files
import os
import zipfile
import pandas as pd
import random
import string
from io import BytesIO
import openai  # Add this line to import the OpenAI module

# Function to generate announcement code
def generate_announcement_code(file_name):
    return file_name[:5]

# Function to process a CSV file
def process_csv_file(file_content, file_name):
    df = pd.read_csv(BytesIO(file_content))
    df['Announcements_Code'] = generate_announcement_code(file_name)
    return df

# Function to process a zip file
def process_zip(zip_content):
    with zipfile.ZipFile(BytesIO(zip_content), 'r') as zip_ref:
        output_zip_buffer = BytesIO()
        with zipfile.ZipFile(output_zip_buffer, 'w') as new_zip_ref:
            for file_name in zip_ref.namelist():
                with zip_ref.open(file_name) as file:
                    df = process_csv_file(file.read(), os.path.splitext(file_name)[0])
                    new_file_name = f"{os.path.splitext(file_name)[0]}_processed.csv"
                    new_csv_path = os.path.join(new_file_name)
                    df.to_csv(new_csv_path, index=False)
                    new_zip_ref.write(new_csv_path, arcname=new_file_name)
        return output_zip_buffer.getvalue()

# Upload 5 zip files
uploaded_files = files.upload()

# Process each uploaded zip file
combined_summaries_df = pd.DataFrame(columns=['Announcements_Code', 'Summary'])

for uploaded_file_name, zip_content in uploaded_files.items():
    processed_zip_content = process_zip(zip_content)

    # Save the processed zip file
    output_zip_path = f'/content/Code_processed_output_{uploaded_file_name}.zip'

    with open(output_zip_path, 'wb') as output_zip_file:
        output_zip_file.write(processed_zip_content)

    # Assuming you have the OpenAI API key set up
    openai.api_key = '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'

    # Process the generated zip file
    with zipfile.ZipFile(BytesIO(processed_zip_content), 'r') as processed_zip_ref:
        for file_name in processed_zip_ref.namelist():
            if file_name.endswith('_processed.csv'):
                with processed_zip_ref.open(file_name) as file:
                    df = pd.read_csv(file)
                    column_to_analyze = 'Content_Label'
                    prompt = f"You are a helpful assistant. Provide a phrase of not more than 100 words to capture the latent meaning in the following terms delimited as context, as these are discord channel discussions from an NFT community named pudgy penguin:'''{df[column_to_analyze].values[0]}'''"

                    # OpenAI API call
                    response = openai.ChatCompletion.create(
                        model="gpt-3.5-turbo-16k",
                        messages=[
                            {"role": "system", "content": "You are a helpful assistant."},
                            {"role": "user", "content": prompt}
                        ]
                    )
                    summary = response['choices'][0]['message']['content']

                    # Extract Announcement code from the CSV file
                    Announcements_Code = df['Announcements_Code'].unique()[0]

                    # Save the summary to the combined_summaries_df DataFrame
                    combined_summaries_df = combined_summaries_df.append({
                        'Announcements_Code': Announcements_Code,
                        'Summary': summary
                    }, ignore_index=True)

# Save the combined summaries to a single CSV file
combined_summaries_df.to_csv('/content/All_summaries_Code_processed_output_AllPhases.csv', index=False)

print("Summaries have been generated and saved in 'All_summaries_Code_processed_output_AllPhases.csv'.")
```
