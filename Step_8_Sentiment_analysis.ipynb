```python
import pandas as pd
from textblob import TextBlob

# Read the csv file
df = pd.read_csv('merged_and_cleaned Phase1forBertopic_data1.csv')

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Perform sentiment analysis and store the results in a new column
df['sentiment'] = df['Content'].apply(lambda x: TextBlob(str(x)).sentiment.polarity)

# Group by month and calculate the average sentiment
monthly_sentiment = df.groupby(df['Date'].dt.to_period('M'))['sentiment'].mean().reset_index()

# Save the aggregated data to a new Excel file
monthly_sentiment.to_excel('monthly_sentiment_analysis.xlsx', index=False)

# View the first 10 rows of the aggregated DataFrame
print(monthly_sentiment.head(10))
```
